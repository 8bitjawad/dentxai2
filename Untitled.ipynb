```python
import os, random, shutil

# Paths
image_dir = r"C:\Users\Abdullah\intraoral\images"
label_dir = r"C:\Users\Abdullah\intraoral\labels"
output_dir = r"C:\Users\Abdullah\intraoral"

# Ratios
train_ratio = 0.7
val_ratio = 0.2
test_ratio = 0.1  # remaining

# Get all images
image_exts = (".jpg", ".jpeg", ".png")
images = [f for f in os.listdir(image_dir) if f.lower().endswith(image_exts)]
random.shuffle(images)

# Split sizes
n_total = len(images)
n_train = int(n_total * train_ratio)
n_val = int(n_total * val_ratio)
# Whatever is left goes to test
n_test = n_total - n_train - n_val

train_files = images[:n_train]
val_files = images[n_train:n_train + n_val]
test_files = images[n_train + n_val:]

def move_files(files, split):
    img_out = os.path.join(output_dir, "images", split)
    lbl_out = os.path.join(output_dir, "labels", split)
    os.makedirs(img_out, exist_ok=True)
    os.makedirs(lbl_out, exist_ok=True)

    for f in files:
        # Copy image
        shutil.copy(os.path.join(image_dir, f), os.path.join(img_out, f))
        # Copy label
        label_file = os.path.splitext(f)[0] + ".txt"
        lbl_path = os.path.join(label_dir, label_file)
        if os.path.exists(lbl_path):
            shutil.copy(lbl_path, os.path.join(lbl_out, label_file))

# Move files
move_files(train_files, "train")
move_files(val_files, "val")
move_files(test_files, "test")

print(f"Split {n_total} images into:")
print(f" - Train: {len(train_files)}")
print(f" - Val:   {len(val_files)}")
print(f" - Test:  {len(test_files)}")
```

```python
# from ultralytics import YOLO
# from easy_explain import YOLOv8LRP
# import cv2
# import torch
# import matplotlib.pyplot as plt

# # Step 1: Load your YOLOv8 model
# model_path = r"C:\dentxai2\DentXAI\runs\detect\intraoral\weights\best.pt"
# model = YOLO(model_path)

# # Step 2: Set image path
# image_path = r"C:\dentxai2\DentXAI\test1.jpeg"

# # Step 3: Run a prediction first to see what's detected
# results = model.predict(image_path, save=False, verbose=False)
# print(f"Model classes: {model.names}")
# print(f"Detections: {len(results[0].boxes)} objects detected")

# # Print detected classes in this image
# if len(results[0].boxes) > 0:
#     detected_classes = results[0].boxes.cls.cpu().numpy()
#     print(f"Detected class indices: {detected_classes}")
#     for cls_idx in detected_classes:
#         print(f"  - {model.names[int(cls_idx)]}")

# # Step 4: Load image using OpenCV and convert to tensor
# image = cv2.imread(image_path)
# image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# image = cv2.resize(image, (640, 640))

# # Convert to tensor and normalize
# image_tensor = torch.from_numpy(image).permute(2, 0, 1).float() / 255.0

# # Step 5: Initialize YOLOv8LRP
# try:
#     lrp = YOLOv8LRP(model.model, power=2, eps=1, device='cpu')
# except:
#     lrp = YOLOv8LRP(model, power=2, eps=1, device='cpu')

# # Step 6: Generate explanations for ALL classes
# # Your classes should be: 0=low, 1=moderate, 2=severe
# num_classes = len(model.names)

# # Option A: Explain ALL classes
# print("\n" + "="*50)
# print("Generating XAI explanations for all classes...")
# print("="*50)

# for class_idx in range(num_classes):
#     class_name = model.names[class_idx]
#     print(f"\nGenerating explanation for class {class_idx}: {class_name}")
    
#     try:
#         # Generate explanation for this class
#         explanation_lrp = lrp.explain(image_tensor, cls=class_idx, contrastive=False).cpu()
        
#         # Plot explanation
#         lrp.plot_explanation(
#             frame=image_tensor,
#             explanation=explanation_lrp,
#             contrastive=False,
#             cmap='seismic',
#             title=f'XAI Explanation: {class_name.upper()} Severity'
#         )
#         plt.savefig(f'xai_explanation_{class_name}.png', dpi=150, bbox_inches='tight')
#         print(f"  ✓ Saved: xai_explanation_{class_name}.png")
        
#     except Exception as e:
#         print(f"  ✗ Error generating explanation for {class_name}: {e}")

# # Option B: Explain ONLY detected classes
# print("\n" + "="*50)
# print("Generating XAI explanations for DETECTED classes only...")
# print("="*50)

# if len(results[0].boxes) > 0:
#     unique_detected_classes = set(results[0].boxes.cls.cpu().numpy().astype(int))
    
#     for class_idx in unique_detected_classes:
#         class_name = model.names[class_idx]
#         print(f"\nGenerating explanation for detected class: {class_name}")
        
#         try:
#             explanation_lrp = lrp.explain(image_tensor, cls=class_idx, contrastive=False).cpu()
            
#             lrp.plot_explanation(
#                 frame=image_tensor,
#                 explanation=explanation_lrp,
#                 contrastive=False,
#                 cmap='seismic',
#                 title=f'XAI Explanation: {class_name.upper()} (Detected)'
#             )
#             plt.savefig(f'xai_detected_{class_name}.png', dpi=150, bbox_inches='tight')
#             print(f"  ✓ Saved: xai_detected_{class_name}.png")
            
#         except Exception as e:
#             print(f"  ✗ Error: {e}")
# else:
#     print("No objects detected in the image!")

# # Option C: Compare explanations side-by-side
# print("\n" + "="*50)
# print("Creating comparison visualization...")
# print("="*50)

# fig, axes = plt.subplots(1, num_classes, figsize=(15, 5))

# for idx, class_idx in enumerate(range(num_classes)):
#     class_name = model.names[class_idx]
    
#     try:
#         explanation_lrp = lrp.explain(image_tensor, cls=class_idx, contrastive=False).cpu()
        
#         # Convert explanation to numpy for plotting
#         exp_numpy = explanation_lrp.numpy()
        
#         axes[idx].imshow(exp_numpy, cmap='seismic')
#         axes[idx].set_title(f'{class_name.upper()}')
#         axes[idx].axis('off')
        
#     except Exception as e:
#         axes[idx].text(0.5, 0.5, f'Error:\n{str(e)}', 
#                       ha='center', va='center', transform=axes[idx].transAxes)
#         axes[idx].axis('off')

# plt.suptitle('XAI Comparison: All Severity Levels', fontsize=14, fontweight='bold')
# plt.tight_layout()
# plt.savefig('xai_comparison_all_classes.png', dpi=150, bbox_inches='tight')
# print("✓ Saved: xai_comparison_all_classes.png")

# plt.show()

# print("\n" + "="*50)
# print("XAI Generation Complete!")
# print("="*50)

from ultralytics import YOLO
from easy_explain import YOLOv8LRP
import cv2
import torch
import matplotlib.pyplot as plt

# Step 1: Load your YOLOv8 model
model_path = r"C:\dentxai2\DentXAI\runs\detect\intraoral\weights\best.pt"
model = YOLO(model_path)

# Step 2: Set image path
image_path = r"C:\dentxai2\DentXAI\test2.jpg"

# Step 3: Run a prediction first to see what's detected
results = model.predict(image_path, save=False, verbose=False)
print(f"Model classes: {model.names}")
print(f"Detections: {len(results[0].boxes)} objects detected")

# Print detected classes in this image
if len(results[0].boxes) > 0:
    detected_classes = results[0].boxes.cls.cpu().numpy()
    print(f"Detected class indices: {detected_classes}")
    for cls_idx in detected_classes:
        print(f"  - {model.names[int(cls_idx)]}")

# Step 4: Load image using OpenCV and convert to tensor
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize(image, (640, 640))

# Convert to tensor and normalize
image_tensor = torch.from_numpy(image).permute(2, 0, 1).float() / 255.0

# Step 5: Initialize YOLOv8LRP
try:
    lrp = YOLOv8LRP(model.model, power=2, eps=1, device='cpu')
except:
    lrp = YOLOv8LRP(model, power=2, eps=1, device='cpu')

# Step 6: Generate explanations for ALL classes
# Your classes should be: 0=low, 1=moderate, 2=severe
num_classes = len(model.names)

# Option A: Explain ALL classes
print("\n" + "="*50)
print("Generating XAI explanations for all classes...")
print("="*50)

for class_idx in range(num_classes):
    class_name = model.names[class_idx]
    print(f"\nGenerating explanation for class {class_idx}: {class_name}")
    
    try:
        # Generate explanation for this class
        explanation_lrp = lrp.explain(image_tensor, cls=class_idx, contrastive=False).cpu()
        
        # Plot explanation
        lrp.plot_explanation(
            frame=image_tensor,
            explanation=explanation_lrp,
            contrastive=False,
            cmap='seismic',
            title=f'XAI Explanation: {class_name.upper()} Severity'
        )
        plt.savefig(f'xai_explanation_{class_name}.png', dpi=150, bbox_inches='tight')
        print(f"  ✓ Saved: xai_explanation_{class_name}.png")
        
    except Exception as e:
        print(f"  ✗ Error generating explanation for {class_name}: {e}")

# Option B: Explain ONLY detected classes
print("\n" + "="*50)
print("Generating XAI explanations for DETECTED classes only...")
print("="*50)

if len(results[0].boxes) > 0:
    unique_detected_classes = set(results[0].boxes.cls.cpu().numpy().astype(int))
    
    for class_idx in unique_detected_classes:
        class_name = model.names[class_idx]
        print(f"\nGenerating explanation for detected class: {class_name}")
        
        try:
            explanation_lrp = lrp.explain(image_tensor, cls=class_idx, contrastive=False).cpu()
            
            lrp.plot_explanation(
                frame=image_tensor,
                explanation=explanation_lrp,
                contrastive=False,
                cmap='seismic',
                title=f'XAI Explanation: {class_name.upper()} (Detected)'
            )
            plt.savefig(f'xai_detected_{class_name}.png', dpi=150, bbox_inches='tight')
            print(f"  ✓ Saved: xai_detected_{class_name}.png")
            
        except Exception as e:
            print(f"  ✗ Error: {e}")
else:
    print("No objects detected in the image!")

# Option C: Compare explanations side-by-side WITH CONFIDENCE SCORES
print("\n" + "="*50)
print("Creating comparison visualization with confidence analysis...")
print("="*50)

# First, get prediction confidence for each class
raw_results = model.predict(image_path, save=False, verbose=False)[0]
boxes = raw_results.boxes

# Calculate average confidence per class
class_confidences = {i: [] for i in range(num_classes)}
for box in boxes:
    cls_idx = int(box.cls.item())
    conf = float(box.conf.item())
    class_confidences[cls_idx].append(conf)

# Compute average confidence
avg_confidence = {}
for cls_idx in range(num_classes):
    if class_confidences[cls_idx]:
        avg_confidence[cls_idx] = sum(class_confidences[cls_idx]) / len(class_confidences[cls_idx])
    else:
        avg_confidence[cls_idx] = 0.0

print("\n📊 Class Confidence Analysis:")
for cls_idx in range(num_classes):
    class_name = model.names[cls_idx]
    conf = avg_confidence[cls_idx]
    count = len(class_confidences[cls_idx])
    print(f"  {class_name.upper()}: {conf:.3f} confidence ({count} detections)")

# Create comparison plot
fig, axes = plt.subplots(2, num_classes, figsize=(15, 10))

for idx, class_idx in enumerate(range(num_classes)):
    class_name = model.names[class_idx]
    
    try:
        explanation_lrp = lrp.explain(image_tensor, cls=class_idx, contrastive=False).cpu()
        
        # Convert explanation to numpy for plotting
        exp_numpy = explanation_lrp.numpy()
        
        # Top row: Original image with overlay
        axes[0, idx].imshow(image)
        axes[0, idx].imshow(exp_numpy, cmap='seismic', alpha=0.5)
        conf_text = f"Conf: {avg_confidence[class_idx]:.2f}" if avg_confidence[class_idx] > 0 else "Not detected"
        axes[0, idx].set_title(f'{class_name.upper()}\n{conf_text}', fontsize=10, fontweight='bold')
        axes[0, idx].axis('off')
        
        # Bottom row: Pure heatmap
        im = axes[1, idx].imshow(exp_numpy, cmap='seismic')
        axes[1, idx].set_title(f'Heatmap: {class_name}', fontsize=9)
        axes[1, idx].axis('off')
        
        # Add colorbar for the last subplot
        if idx == num_classes - 1:
            plt.colorbar(im, ax=axes[1, idx], fraction=0.046, pad=0.04)
        
    except Exception as e:
        axes[0, idx].text(0.5, 0.5, f'Error:\n{str(e)}', 
                         ha='center', va='center', transform=axes[0, idx].transAxes, fontsize=8)
        axes[0, idx].axis('off')
        axes[1, idx].axis('off')

plt.suptitle('XAI Comparison: All Severity Levels\n(Top: Overlay | Bottom: Heatmap)', 
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('xai_comparison_all_classes.png', dpi=150, bbox_inches='tight')
print("✓ Saved: xai_comparison_all_classes.png")

plt.show()

print("\n" + "="*50)
print("XAI Generation Complete!")
print("="*50)
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from ultralytics import YOLO

# Paths
image_path = r"C:\dentxai2\DentXAI\test1.jpeg"
model_weights = r"C:\dentxai2\DentXAI\runs\detect\intraoral\weights\best.pt"

# Load image and model
image = np.array(Image.open(image_path).convert('RGB'))
model = YOLO(model_weights)
model.fuse()

print(f"Model classes: {model.names}")
print("="*60)

# Original prediction
orig_results = model(image)[0]

# Get original confidence per class
orig_conf_per_class = {}
for i in range(len(model.names)):
    # Filter boxes for this class
    class_boxes = [box for box in orig_results.boxes if int(box.cls) == i]
    if class_boxes:
        orig_conf_per_class[i] = max([float(box.conf) for box in class_boxes])
    else:
        orig_conf_per_class[i] = 0.0

print("Original detections:")
for cls_idx, conf in orig_conf_per_class.items():
    class_name = model.names[cls_idx]
    print(f"  {class_name}: {conf:.3f} confidence")
print("="*60)

# Occlusion parameters
patch_size = (50, 50)
step_size = 25
height, width, _ = image.shape

# Prepare heatmap for EACH class
heatmaps = {i: np.zeros((height, width), dtype=np.float32) for i in range(len(model.names))}

# Occlusion loop
print("Generating occlusion maps...")
total_patches = ((height - patch_size[1]) // step_size + 1) * ((width - patch_size[0]) // step_size + 1)
patch_count = 0

for y in range(0, height - patch_size[1] + 1, step_size):
    for x in range(0, width - patch_size[0] + 1, step_size):
        occluded = image.copy()
        region = occluded[y:y+patch_size[1], x:x+patch_size[0]]
        blurred = cv2.GaussianBlur(region, (21, 21), 0)
        occluded[y:y+patch_size[1], x:x+patch_size[0]] = blurred
        
        # Run model on occluded image
        occl_results = model(occluded)[0]
        
        # Get confidence per class for occluded image
        occl_conf_per_class = {}
        for i in range(len(model.names)):
            class_boxes = [box for box in occl_results.boxes if int(box.cls) == i]
            if class_boxes:
                occl_conf_per_class[i] = max([float(box.conf) for box in class_boxes])
            else:
                occl_conf_per_class[i] = 0.0
        
        # Calculate difference for each class
        for cls_idx in range(len(model.names)):
            diff = float(orig_conf_per_class[cls_idx] - occl_conf_per_class[cls_idx])
            heatmaps[cls_idx][y:y+patch_size[1], x:x+patch_size[0]] += diff
        
        patch_count += 1
        if patch_count % 50 == 0:
            print(f"  Processed {patch_count}/{total_patches} patches...")

print(f"✓ Completed {patch_count} patches")
print("="*60)

# Normalize each heatmap
for cls_idx in range(len(model.names)):
    hmap = heatmaps[cls_idx]
    if hmap.max() > hmap.min():
        heatmaps[cls_idx] = (hmap - hmap.min()) / (hmap.max() - hmap.min())
    else:
        heatmaps[cls_idx] = np.zeros_like(hmap)

# OPTION 1: Individual heatmaps for each class
print("Generating individual heatmaps...")
for cls_idx in range(len(model.names)):
    class_name = model.names[cls_idx]
    conf = orig_conf_per_class[cls_idx]
    
    plt.figure(figsize=(10, 6))
    plt.imshow(image)
    plt.imshow(heatmaps[cls_idx], cmap='jet', alpha=0.5)
    plt.title(f'Occlusion Heatmap: {class_name.upper()}\nOriginal Confidence: {conf:.3f}')
    plt.axis('off')
    plt.colorbar(label='Impact on Detection', fraction=0.046, pad=0.04)
    plt.savefig(f'occlusion_heatmap_{class_name}.png', dpi=150, bbox_inches='tight')
    print(f"  ✓ Saved: occlusion_heatmap_{class_name}.png")
    plt.close()

# OPTION 2: Side-by-side comparison
print("Generating comparison visualization...")
num_classes = len(model.names)
fig, axes = plt.subplots(2, num_classes, figsize=(15, 10))

for idx in range(num_classes):
    class_name = model.names[idx]
    conf = orig_conf_per_class[idx]
    
    # Top row: Image with overlay
    axes[0, idx].imshow(image)
    axes[0, idx].imshow(heatmaps[idx], cmap='jet', alpha=0.5)
    axes[0, idx].set_title(f'{class_name.upper()}\nConf: {conf:.3f}', 
                           fontsize=10, fontweight='bold')
    axes[0, idx].axis('off')
    
    # Bottom row: Pure heatmap
    im = axes[1, idx].imshow(heatmaps[idx], cmap='jet')
    axes[1, idx].set_title(f'Heatmap Only', fontsize=9)
    axes[1, idx].axis('off')
    
    # Add colorbar to last subplot
    if idx == num_classes - 1:
        plt.colorbar(im, ax=axes[1, idx], fraction=0.046, pad=0.04)

plt.suptitle('Occlusion Sensitivity Analysis: All Severity Levels\n(Brighter = More Important)', 
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('occlusion_comparison_all_classes.png', dpi=150, bbox_inches='tight')
print("  ✓ Saved: occlusion_comparison_all_classes.png")

# OPTION 3: Combined heatmap showing which class is most affected by each region
print("Generating combined importance map...")
plt.figure(figsize=(12, 8))

# Create RGB composite where each class gets a color channel
combined_heatmap = np.zeros((height, width, 3), dtype=np.float32)
colors = [
    [1, 0, 0],  # Red for class 0 (low)
    [0, 1, 0],  # Green for class 1 (moderate)
    [0, 0, 1]   # Blue for class 2 (severe)
]

for cls_idx in range(min(num_classes, 3)):  # Limit to 3 for RGB
    combined_heatmap[:, :, cls_idx] = heatmaps[cls_idx]

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(image)
plt.imshow(combined_heatmap, alpha=0.6)
plt.title('Combined Occlusion Map\n(R=Low, G=Moderate, B=Severe)')
plt.axis('off')

plt.tight_layout()
plt.savefig('occlusion_combined_rgb.png', dpi=150, bbox_inches='tight')
print("  ✓ Saved: occlusion_combined_rgb.png")

plt.show()

print("="*60)
print("✅ Occlusion analysis complete!")
print("\nInterpretation:")
print("  • Bright areas = Occluding this region strongly reduces detection")
print("  • Dark areas = Less important for detection")
print("  • Compare across classes to see severity-specific features")
print("="*60)
```
